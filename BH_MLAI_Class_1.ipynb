## 4. Handling Missing Data

```python
# Adding missing values to rows 4 and 5 in the 'Age' and 'City' columns
df.loc[4, 'Age'] = None
df.loc[5, 'City'] = None
df.loc[6, 'Salary'] = None

print("\nDataFrame with Missing Values:")
print(df)
```

```python
df.isnull().sum()
```

```python
# Dropping rows that contain any missing values
df_dropped = df.dropna()
print("\nDataFrame After Dropping Missing Values:")
print(df_dropped)
```

```python
# Filling missing values with a default value
df_filled = df.fillna('Unknown')
print("\nDataFrame After Filling Missing Values with 'Unknown':")
print(df_filled)
```

```python
# Fill missing values in 'Age' with the mean age, and in 'City' with 'Unknown'
df['Age'] = df['Age'].fillna(df['Age'].mean())
df['City'] = df['City'].fillna('Unknown')
df['Salary'] = df['Salary'].fillna(df['Salary'].median())

print("\nDataFrame After Filling Missing Values with Specific Values per Column:")
print(df)
```

## 6. Merging and Joining DataFrames

```python
import pandas as pd

# Original DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva', 'Frank', 'Grace', 'Hannah'],
    'Age': [24, 27, 22, 32, 28, 26, 29, 31],
    'City': ['New York', 'San Francisco', 'Chicago', 'Los Angeles', 'Seattle', 'Austin', 'Denver', 'Boston'],
    'Profession': ['Engineer', 'Doctor', 'Artist', 'Lawyer', 'Architect', 'Designer', 'Chef', 'Writer'],
    'Salary': [85000, 95000, 70000, 120000, 105000, 78000, 68000, 74000]
}

df = pd.DataFrame(data)

# Second DataFrame with City and State information
data_cities = {
    'City': ['New York', 'San Francisco', 'Chicago', 'Los Angeles', 'Seattle', 'Austin', 'Boston', 'Miami'],
    'State': ['NY', 'CA', 'IL', 'CA', 'WA', 'TX', 'MA', 'FL']
}

df_cities = pd.DataFrame(data_cities)
```

Inner Join
An inner join returns rows where there is a match in both DataFrames. Here, we’ll join df and df_cities on the City column, keeping only the rows where cities are in both DataFrames.

```python
# Inner join on 'City'
df_inner = pd.merge(df, df_cities, on='City', how='inner')
print("Inner Join Result:")
print(df_inner)
```

```python
# Left join on 'City'
df_left = pd.merge(df, df_cities, on='City', how='left')
print("\nLeft Join Result:")
print(df_left)
```

## 7. Data Transformation and Applying Functions (skip this part if running out of time)

```python
import pandas as pd

# Sample DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva', 'Frank', 'Grace', 'Hannah'],
    'Age': [24, 27, 22, 32, 28, 26, 29, 31],
    'City': ['New York', 'San Francisco', 'Chicago', 'Los Angeles', 'Seattle', 'Austin', 'Denver', 'Boston'],
    'Profession': ['Engineer', 'Doctor', 'Artist', 'Lawyer', 'Architect', 'Designer', 'Chef', 'Writer'],
    'Salary': [85000, 95000, 70000, 120000, 105000, 78000, 68000, 74000]
}

df = pd.DataFrame(data)

# 1. Applying a function to a column
# Example: Increase each salary by 10%
df['Salary10'] = df['Salary'].apply(lambda x: x * 1.10)
print("Updated Salaries with 10% increase:")
print(df[['Name', 'Salary','Salary10']])
```

```python
# 2. Creating a new column based on a condition
# Example: Adding a column to categorize people as 'Young' or 'Mature' based on age
df['Age_Group'] = df['Age'].apply(lambda x: 'Young' if x < 30 else 'Mature')
print("\nDataFrame with Age Group column:")
print(df[['Name', 'Age', 'Age_Group']])
```

```python
# 3. Using `map` to transform data based on a dictionary
# Example: Mapping cities to regions (East or West)
city_to_region = {
    'New York': 'East',
    'San Francisco': 'West',
    'Chicago': 'Midwest',
    'Los Angeles': 'West',
    'Seattle': 'West',
    'Austin': 'South',
    'Denver': 'West',
    'Boston': 'East'
}
df['Region'] = df['City'].map(city_to_region)
print("\nDataFrame with Region column based on City:")
print(df[['Name', 'City', 'Region']])
```

```python
# 4. Creating a new column with complex calculations
# Example: Calculate years until retirement assuming retirement age is 65
df['Years_to_Retire'] = df['Age'].apply(lambda x: 65 - x)
print("\nDataFrame with Years to Retirement column:")
print(df[['Name', 'Age', 'Years_to_Retire']])
```

```python
# 5. Using `apply` on rows (axis=1)
# Example: Combine Name and Profession into a single column
df['Name_Profession'] = df.apply(lambda row: f"{row['Name']} - {row['Profession']}", axis=1)
print("\nDataFrame with combined Name and Profession column:")
print(df[['Name', 'Profession', 'Name_Profession']])
```

```python
from sqlalchemy import create_engine
import pandas as pd

# Create an engine with SQLAlchemy
engine = create_engine("mysql+mysqlconnector://root:password@localhost:3306/PRACTICE_DEV")

# Define your query
query = """
SELECT 
    p.product_id,
    p.product_name,
    s.supplier_name,
    l.location_name,
    SUM(so.quantity) AS total_sales_quantity,
    SUM(po.quantity) AS total_purchase_quantity,
    AVG(po.quantity) AS average_purchase_quantity,
    MAX(i.quantity) AS max_inventory_quantity,
    MIN(i.quantity) AS min_inventory_quantity
FROM 
    products p
LEFT JOIN 
    inventory i ON p.product_id = i.product_id
LEFT JOIN 
    locations l ON i.location_id = l.location_id
LEFT JOIN 
    suppliers s ON p.product_id = s.supplier_id
LEFT JOIN 
    sales_orders so ON p.product_id = so.product_id AND so.location_id = l.location_id
LEFT JOIN 
    purchase_orders po ON p.product_id = po.product_id AND po.supplier_id = s.supplier_id
WHERE 
    l.location_name IN ('Warehouse C', 'Warehouse B')
    AND p.product_name IN ('Product E', 'Product B','Product I')    
GROUP BY 
    p.product_id, l.location_id
HAVING 
    total_sales_quantity > 800
ORDER BY 
    total_sales_quantity DESC;
"""
# Use pandas with the SQLAlchemy engine
df = pd.read_sql(query, engine)

# Display the dataframe
from IPython.display import display

# Display the DataFrame in a nicer format
display(df)
```

In the .plot() method of pandas, some of the most commonly used parameters (or arguments) allow you to customize the appearance and behavior of your plots. Here’s a list of the most popular parameters with brief explanations:

1. kind
Specifies the type of plot. Common values:
'line': Line plot (default)
'bar': Vertical bar plot
'barh': Horizontal bar plot
'hist': Histogram
'box': Box plot
'scatter': Scatter plot (requires x and y parameters)
'pie': Pie chart
Example: df.plot(kind='bar')
2. x and y
Specify the column names to use for the x-axis and y-axis, respectively.
Only relevant for certain types of plots, like scatter or line.
Example: df.plot(kind='scatter', x='Column1', y='Column2')
3. title
Sets the title of the plot.
Example: df.plot(kind='bar', title='My Bar Plot')
4. figsize
Controls the size of the figure in inches. Takes a tuple, e.g., (width, height).
Example: df.plot(kind='bar', figsize=(10, 6))
5. color
Sets the color of the plot elements (like bars, lines, etc.).
Can be a single color or a list of colors if you want to color different elements.
Example: df.plot(kind='bar', color='skyblue')
6. legend
Controls the display of the legend. True to show, False to hide.
Example: df.plot(kind='line', legend=False)
7. grid
Adds a grid to the plot if set to True.
Example: df.plot(kind='line', grid=True)
8. xlabel and ylabel
Label the x-axis and y-axis.
Example: df.plot(kind='bar', xlabel='Categories', ylabel='Values')
9. rot
Rotates the labels on the x-axis. Commonly used for bar plots to prevent label overlap.
Example: df.plot(kind='bar', rot=45)

### **Principles of Statistics**
* Central Tendency
* Distribucion
* Variability (Dispersion)
* Correlation

1. Central Tendency
Central tendency involves measures like the mean, median, and mode, which represent the "center" of the data. Let’s calculate these for relevant columns like Age and Salary.

```python
# Mean, Median, and Mode for Age and Salary
central_tendency = {
    'Mean Age': nbadata['Age'].mean(),
    'Median Age': nbadata['Age'].median(),
    'Mode Age': nbadata['Age'].mode()[0],
    'Mean Salary': nbadata['Salary'].mean(),
    'Median Salary': nbadata['Salary'].median(),
    'Mode Salary': nbadata['Salary'].mode()[0]
}

print("Central Tendency Measures:")
print(central_tendency)
```

2. Distribution
The distribution shows how data points are spread across values. For this, you can use a histogram to visualize the distribution of Age and Salary.

```python
# Histogram for Age
nbadata['Age'].plot(kind='hist', bins=10, title='Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

# Histogram for Salary
nbadata['Salary'].plot(kind='hist', bins=15, title='Distribution of Salary')
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Plotting histogram with custom bar width and spacing
ax = nbadata['Age'].plot(kind='hist', bins=10, edgecolor='black', width=1.9, title='Distribution of Age with Bar Spacing')
plt.xlabel('Age')
plt.ylabel('Frequency')

plt.show()
```

3. Variability (Dispersion)
Variability or dispersion shows how spread out the data is, using measures like variance and standard deviation.
* Variance: Measures the average squared deviation from the mean.
* Standard Deviation: Square root of variance, indicating the average distance from the mean.

```python
# Variance and Standard Deviation for Age and Salary
variability = {
    'Age Variance': nbadata['Age'].var(),
    'Age Standard Deviation': nbadata['Age'].std(),
    'Salary Variance': nbadata['Salary'].var(),
    'Salary Standard Deviation': nbadata['Salary'].std()
}

print("Variability Measures:")
print(variability)
```

4. Correlation
Correlation measures the strength and direction of the relationship between two variables. In this case, you might be interested in the correlation between Age and Salary, or other pairs of numerical variables.

```python
# Correlation between numerical columns
correlation_matrix = nbadata[['Age', 'Weight', 'Salary']].corr()
print("Correlation Matrix:")
print(correlation_matrix)

# Heatmap for visualization (optional, requires seaborn)
import seaborn as sns
import matplotlib.pyplot as plt

plt.title('Correlation Heatmap')
sns.heatmap(correlation_matrix, annot=True, cmap='Purples', center=0)
plt.show()
```
